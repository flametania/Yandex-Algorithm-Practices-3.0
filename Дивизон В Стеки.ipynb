## Задание 11

```python
stack_lst = []
command = input()
while command != "exit":
  # print(command)
  if command[:4:] == "push":
    stack_lst.append(int(command[5::]))
    print("ok")
  elif command == "pop":
    if len(stack_lst) == 0:
      print("error")
    else:
      print(stack_lst[len(stack_lst) - 1])
      stack_lst.pop()
  elif command == "back":
    if len(stack_lst) == 0:
      print("error")
    else:
      print(stack_lst[len(stack_lst) - 1])
  elif command == "size":
    print(len(stack_lst))
  elif command == "clear":
    stack_lst.clear()
    print("ok")
  command = input()
if command == "exit":
  print("bye")
```

## Задание 12

```python
stack_lst = []
seq = input()
open_brackets = "({["
close_brackets_dict = {")": "(", "}": "{", "]": "["}
for s in seq:
  if s in open_brackets:
    stack_lst.append(s)
  else:
    if len(stack_lst) > 0 and close_brackets_dict[s] == stack_lst[len(stack_lst) - 1]:
      stack_lst.pop()
    else:
      stack_lst.append("e")
      break
if len(stack_lst) == 0:
  print("yes")
else:
  print("no")
```

## Задание 13

```python
digits = "0123456789"
postfix_lst = input().split()
res = 0
stack_lst = []
for op in postfix_lst:
  if op in digits:
    stack_lst.append(int(op))
  else:
    if op == "+":
      res = stack_lst[-1] + stack_lst[-2]
    elif op == "-":
      res = stack_lst[-2] - stack_lst[-1]
    elif op == "*":
      res = stack_lst[-1] * stack_lst[-2]
    stack_lst.pop()
    stack_lst.pop()
    stack_lst.append(res)
print(stack_lst[0])
```

## 15

```python
n = int(input())
lst = list(map(int, input().split()))

# n = 10
# lst = [1, 2, 3, 2, 1, 4, 2, 5, 3, 1]
stack_lst = []
answer_lst = [-1 for _ in range(n)]
for i in range(n):
  while len(stack_lst) > 0 and lst[i] < stack_lst[-1][0]:
    answer_lst[stack_lst.pop()[1]] = i
  stack_lst.append([lst[i], i])
print(" ".join(map(str, answer_lst)))
```
