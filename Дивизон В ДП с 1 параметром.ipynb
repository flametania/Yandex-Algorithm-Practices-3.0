## 21

```python
n, k = map(int, input().split())
dp = [0] * (n + 1)
dp[0] = dp[1] = 1
i = 1
for i in range(2, n + 1):
  if i < k:
    dp[i] = sum(dp[:i:])
  else:
    dp[i] = sum(dp[i - k:i:])
  # print(dp)
print(dp[i - 1])
```

## 23

```python
n = int(input())
dp, prev = [0] * (n + 1), [0] * (n + 1)
dp[0], prev[0] = 0, 1
for i in range(2, n + 1):
  variants = []
  variants.append((dp[i - 1] + 1, i - 1))
  if i % 2 == 0:
    variants.append((dp[i // 2] + 1, i // 2))
  if i % 3 == 0:
    variants.append((dp[i // 3] + 1, i // 3))
  dp[i], prev[i] = min(variants, key=lambda x: x[0])
else:
  i = 1
# print(dp, prev)
print(dp[-1])
ans = [n]
step = prev[-1]
while step > 0:
  ans.append(step)
  step = prev[step]
# print(ans)
print(" ".join(map(str, ans[::-1])))
```
