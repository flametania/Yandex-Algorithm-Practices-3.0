```python
n = int(input())
seq = list(map(int, input().split()))
# n = 6
# seq = [3, 29, 5, 5, 28, 6]
# n = 10
# seq = [4, 8, 2, 6, 2, 10, 6, 29, 58, 9]
# n = 7
# seq = [4, 10, 5, 12, 3, 24, 7]
# n = 21
# seq = [10, -10, 9, -9, 8, -8, 7, -7, 6, -6, 5, -5, 4, -4, 3, -3, 2, -2, 1, -1, 0]
dp = [1] * (n + 1)
prev = [-1] * (n + 1)
# dp[0] = 1
# prev[0] = -1
max_subseq_len = 0
max_subseq_ind = 0
for i in range(1, n):
  for j in range(0, i):
    if seq[j] < seq[i] and dp[j] + 1 > dp[i]:
      dp[i] = dp[j] + 1
      prev[i] = j
      if dp[i] > max_subseq_len:
        max_subseq_len = dp[i]
        max_subseq_ind = i
# print(dp)
# print(prev)
i = max_subseq_ind
ans_lst = []
while i > -1:
  ans_lst.append(seq[i])
  i = prev[i]
print(" ".join(map(str, ans_lst[::-1])))
```
