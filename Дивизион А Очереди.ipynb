```python
from collections import deque

min_deq = deque()
with open('input.txt') as f_in:
  n, k = map(int, f_in.readline().rstrip('\n').split())
  lst = list(map(int, f_in.readline().rstrip('\n').split()))
# n, k = 7, 3
# lst = list(map(int, "1 3 2 4 5 3 1".split()))

min_lst = [0] * (n - k + 1)
for i in range(k):
  while len(min_deq) > 0 and min_deq[-1] > lst[i]:
    min_deq.pop()
  if len(min_deq) == 0 or min_deq[-1] <= lst[i]:
    min_deq.append(lst[i])
min_lst[0] = min_deq[0]
# print(min_deq[0])
# print(min_deq)

for i in range(1, n - k + 1):
  # print(i, i + k - 1)
  if min_deq[0] == lst[i - 1]:
    min_deq.popleft()
  while len(min_deq) > 0 and min_deq[-1] > lst[i + k - 1]:
    min_deq.pop()
  if len(min_deq) == 0 or min_deq[-1] <= lst[i + k - 1]:
    min_deq.append(lst[i + k - 1])
  min_lst[i] = min_deq[0]
  # print(min_deq[0])
  # print(min_deq, min_lst)
with open('output.txt', 'w') as f_out:
  print("\n".join(map(str, min_lst)), file=f_out)
```
