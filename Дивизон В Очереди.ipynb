```python
import queue

q = queue.Queue()
command = input()
while command != "exit":
  # print(command)
  if command[:4:] == "push":
    q.put(int(command[5::]))
    print("ok")
  elif command == "pop":
    if q.empty():
      print("error")
    else:
      # print(q[len(stack_lst) - 1])
      print(q.get())
  elif command == "front":
    if q.empty():
      print("error")
    else:
      print(q.queue[0])
  elif command == "size":
    print(q.qsize())
  elif command == "clear":
    q.queue.clear()
    print("ok")
  command = input()
if command == "exit":
  print("bye")
```

## Задание 18

```python
from collections import deque

d = deque()
command = input()
while command != "exit":
  # print(command)
  if command[:10:] == "push_front":
    d.appendleft(int(command[11::]))
    print("ok")
  elif command[:9:] == "push_back":
    d.append(int(command[10::]))
    print("ok")
  elif command == "pop_front":
    if len(d) == 0:
      print("error")
    else:
      print(d.popleft())
  elif command == "pop_back":
    if len(d) == 0:
      print("error")
    else:
      print(d.pop())
  elif command == "front":
    if len(d) == 0:
      print("error")
    else:
      print(d[0])
  elif command == "back":
    if len(d) == 0:
      print("error")
    else:
      print(d[-1])
  elif command == "size":
    print(len(d))
  elif command == "clear":
    d.clear()
    print("ok")
  command = input()
if command == "exit":
  print("bye")
```
